# bank_rate 예측하는 알고리즘 만들기
 신용점수가 높을수록 상품 추천을 많이 받을 것이라고 생각하여 만듦

```python
# 필요없는 변수 제거
train_pro = train.drop(['user_id', 'application_count', 'application_id', 'insert_time', 'personal_rehabilitation_yn', 'NA_count', 
                        'birth_year', 'company_enter_month', 'loanapply_insert_time', 'apply_rate', 'is_applied', 'gender'], axis=1)
```

```python
train_pro.info()
```

```python
train_pro['income_type'] = train_pro['income_type'].astype('category')
train_pro['houseown_type'] = train_pro['houseown_type'].astype('category')
train_pro['purpose'] = train_pro['purpose'].astype('category')
train_pro['employment_type'] = train_pro['employment_type'].astype('category')
```

```python
data =train_pro.drop(['product_rate'], axis=1)
label=train_pro[['product_rate']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, label, test_size=0.2, random_state=11)
lgb_train = lgb.Dataset(X_train, y_train)
lgb_eval = lgb.Dataset(X_test, y_test, reference=lgb_train)
```

# test데이터 가져오기

```python
test=pd.read_csv("./test.csv", encoding = 'cp949',  index_col = 0)
test.isna().sum()
```

```python
# 필요없는 변수 제거
test_pro =  test.drop(['user_id', 'application_id', 'insert_time', 'personal_rehabilitation_yn', 'NA_count', 
                        'birth_year', 'company_enter_month', 'loanapply_insert_time', 'is_applied', 'gender'], axis=1)
```

```python
test_pro.info()
```

```python
test_pro['income_type'] = test_pro['income_type'].astype('category')
test_pro['houseown_type'] = test_pro['houseown_type'].astype('category')
test_pro['purpose'] = test_pro['purpose'].astype('category')
test_pro['employment_type'] = test_pro['employment_type'].astype('category')
```

```python
X =test_pro
```

## 파라미터 조절하기

```python
params = {
    'boosting_type': 'gbdt',
    'objective': 'regression',
    'metric': {'l2', 'l1'},
    'num_leaves': 31,
    'learning_rate': 0.1,
    'feature_fraction': 0.8,
    'bagging_fraction': 0.7,
    'bagging_freq': 8,
    'verbose': 0
}

gbm = lgb.train(params,
                lgb_train,
                num_boost_round=600,
                valid_sets=lgb_eval,
                callbacks=[lgb.early_stopping(stopping_rounds=5)])
```

```python
# predict
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration)
# eval
print('The rmse of prediction is:', mean_squared_error(y_test, y_pred) ** 0.5)

print('MAE ;', mean_absolute_error(y_test, y_pred))
```
